```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/PS_20174392719_1491204439457_log.csv')
```

The below column reference:

step: represents a unit of time where 1 step equals 1 hour
type: type of online transaction
amount: the amount of the transaction
nameOrig: customer starting the transaction
oldbalanceOrg: balance before the transaction
newbalanceOrig: balance after the transaction
nameDest: recipient of the transaction
oldbalanceDest: initial balance of recipient before the transaction

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

# Assuming you have a DataFrame named 'df' with your dataset

# Define features (X) and target variable (y)
X =df[['step', 'amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest','type']]

X = pd.get_dummies(data=X, columns=['type'], drop_first=True)

# Define target variable
y = df['isFraud']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(df.shape)
print(X_train.shape[0])
print(X_test.shape[0])

```

NAIVE BAYES

```python

naive_bayes = GaussianNB()
naive_bayes.fit(X_train, y_train)

predictions = naive_bayes.predict(X_test)
```

```python

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
Accuracy_score2=accuracy_score(y_test,predictions)
Precision_score2=precision_score(y_test,predictions)
Recall_score2=recall_score(y_test,predictions)
F1_score2=f1_score(y_test,predictions)
```

Decision Tree

```python

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

# Predict with Decision Tree Classifier
tree_predictions = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
Accuracy_score1=accuracy_score(y_test, tree_predictions)
Precision_score1=precision_score(y_test, tree_predictions)
Recall_score1=recall_score(y_test, tree_predictions)
F1_score1=f1_score(y_test, tree_predictions)
```

```python
# Making predictions
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# Calculate the accuracy
from sklearn.metrics import accuracy_score
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
print('The training accuracy is', train_accuracy)
print('The test accuracy is', test_accuracy)
```

Comparision


```python
import pandas as pd

data = {
    'Metric': ['Accuracy', 'Precision', 'Recall', 'F1-Score'],
    'Decision Tree': [Accuracy_score1, Precision_score1, Recall_score1, F1_score1],
    'naïve bayes': [Accuracy_score2, Precision_score2, Recall_score2, F1_score2]
}

df = pd.DataFrame(data)
print(df)
```

```python
from scipy.stats import ttest_rel

# Assuming you have arrays of evaluation scores for both classifiers (e.g., accuracy_scores1, accuracy_scores2)
t_stat, p_value = ttest_rel(Accuracy_score1,Accuracy_score2)

if p_value < 0.05:
    print("There is a significant difference in performance.")
else:
    print("There is no significant difference in performance.");

```
